## 데이터 불러오기

```python
import pandas as pd

apple = pd.read_csv('exercises/aapl.csv')
amazon = pd.read_csv('exercises/amzn.csv')
google = pd.read_csv('exercises/goog.csv')
netflix = pd.read_csv('exercises/nflx.csv')
facebook = pd.read_csv('exercises/fb.csv')
```

## 각각의 주식데이터 합치기 
- corp 컬럼 생성
- concatenate

```python
new_apple = apple.assign(ticker = 'apple')
new_apple
```

```python
new_google = google.assign(ticker = 'google')
new_google
```

```python
new_amazon = amazon.assign(ticker = 'amazon')
new_amazon
```

```python
new_facebook = facebook.assign(ticker = 'facebook')
new_facebook
```

```python
new_netflix = netflix.assign(ticker = 'netflix')
new_netflix
```

```python
faang_data = pd.concat([new_apple, new_amazon, new_facebook, new_netflix, new_google], axis=0)
faang_data.tail(20)
```

```python
faang_data.to_csv('exercises/faang_data', index = False)
```

```python
faang = pd.read_csv('exercises/faang_data')
faang
```

```python
faang.date = pd.to_datetime(faang.date)
faang.date
```

```python
faang
```

```python
faang = faang.assign(volume = lambda x: x.volume.astype(int))
faang.head(10)
```

```python
faang = faang.sort_values(by=['date', 'ticker'])
faang
```

## faang 데이터의 volume 열에서 가장 낮은 값 7개 찾기

```python
faang.nsmallest(7, 'volume')
```

## melting 하기 (넓은 -> 긴)
## stack하기 인덱스를 여러개로 쌓기
이제 데이터는 긴 형태와 넓은 형태의 중간에 있다.
melt()를 사용해 완전히 긴 형태로 바꾼다. 
- 힌트 : date 와 ticker 는 ID변수다. (이 값들은 각 행을 고유하게 식별할 수 있다.)

- 힌트2 : open, hagh, low, close, volume 열이 분리되지 않도록 나머지 부분도 멜팅 해야 한다.

```python
melt_faang = faang.melt(
    id_vars = ['date', 'ticker'], #식별자
    value_vars = ['open', 'low', 'high', 'close'], #하나의 열로 모아줌(길게 만듬)
    var_name = 'detail', #설명
    value_name = 'value'
)
melt_faang.head(10)
```

```python
melt_faang[melt_faang.date=='2018-07-26']
```

#### **나라이름이 나오는 값들이 object 타입이기 때문에 atype(str)로 스트링으로 변환해주어야 apply 적용할 수 있다.**  

object 타입으로 들어간 컬럼들이 있나 확인

```python
covid.describe(include = object)
```

```python
covid = covid.apply(lambda x: x.astype(str).str.replace('United_States_of_America', 'USA').str.replace('United_Kingdom', 'UK'))
covid
```

```python
filtered_covid = covid[covid.countriesAndTerritories.isin( ['Argentina', 'Brazil', 'China', 'India', 'Colombia', 'Italy', 'Mexico', 
                                           'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'])]
```

#### 데이터 pivot 시키기

**피봇**(pivot) : 긴 형태의 데이터를 **대각선을 중심으로 회전**시켜 넓은 형태의 데이터로 바꾸는 것.(전치 T랑 비슷)

```python
pivoted_covid = filtered_covid.pivot(columns = 'countriesAndTerritories', values = 'cases') #index는 이미 지정되어 있으므로 따로 넣지 않아도 된다.
pivoted_covid
```

```python
filled_covid = pivoted_covid.fillna(0)
```
