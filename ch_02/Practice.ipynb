1. mb 지진 규모 유형으로 사용해 일본 진도의 96번째 백분위수를 찾는다.

```python
import pandas as pd

df = pd.read_csv(
    'data/earthquakes.csv', 
    usecols=['time', 'title', 'place', 'magType', 'mag', 'alert', 'tsunami']
)
```

```python
df[df.magType == 'mb'].describe(percentiles=[0.05, 0.95])
```
