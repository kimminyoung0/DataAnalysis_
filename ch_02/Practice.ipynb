1. mb 지진 규모 유형으로 사용해 일본 진도의 96번째 백분위수를 찾는다.

```python
import pandas as pd

df = pd.read_csv(
    'data/earthquakes.csv', 
    usecols=['time', 'title', 'place', 'magType', 'mag', 'alert', 'tsunami']
)
```

```python
df[df.magType == 'mb'].describe(percentiles=[0.05, 0.96])
```

2. 인도네시아에서 쓰나미가 동반된 지진의 백분율을 구한다

```python
yes_tsunami = len(df[(df['tsunami']==1) & (df['place'].str.contains('ndone'))])
tot_tsunami = len(df[df['place'].str.contains('ndone')])
round(yes_tsunami/tot_tsunami, 2)
```

3. 네바다 지진에 대한 요약 통계를 계산한다.

```python
df.loc[
    (df.place.str.contains(r'NV|Nevada$')),
    ['place', 'alert', 'mag', 'magType', 'title', 'tsunami']
].describe()
```

4. 지진이 불의 고리에 있는 국가나 미국 주에서 발생했는지를 나타내는 열을 추가한다. 알래스카 Alaska, 남극대륙 Antarctica, Antarctic, 볼리비아 Bolivia, 캘리포니아 California, 캐나다 Canada, 칠레 Chile, 코스타리가 Costa Rica, 에콰도르 Ecuador, 피지 Fiji, 과테말라 Guatemala, 인도네시아 Indonesia, 일본 Japan, 카르마덱 제도 Kermadec islands, 멕시코 Mexico, 뉴질랜드 New Zealand, 페루 Peru, 필리핀 Philippins, 러시아 Rusia, 타이완 Taiwan, 통가 Tonga, 워싱턴 Washington을 사용한다.

```python
df_rof = df.assign(
    in_ROF_orUS = df.place.str.contains(r'CA|California|Alaska|Antarctic|Bolivia|Canada|Chile|Costa|Ecuador|Fiji|Guatemala|Indonesia|Japan|Kermadec|Mexico|New Z|Peru|Philippin|Rusia|Taiwan|Tonga|Washington')
)
df_rof
```

```python
df.loc[
    (df.place.str.contains(r'eru')),
    ['place', 'alert', 'mag', 'magType', 'title', 'tsunami']
]
```

5. 불의 고리 위치에서 발생한 지진의 횟수와 그외에서 발생한 지진의 횟수를 계산한다.

```python
print(len(df_rof[df_rof['in_ROF_orUS'] == True]), len(df_rof[df_rof['in_ROF_orUS'] == False]))
```
